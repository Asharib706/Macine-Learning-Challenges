The dataset contains observations of US domestic flights in 2013, and consists of the following fields:

- **Year**: The year of the flight (all records are from 2013)
- **Month**: The month of the flight
- **DayofMonth**: The day of the month on which the flight departed
- **DayOfWeek**: The day of the week on which the flight departed - from 1 (Monday) to 7 (Sunday)
- **Carrier**: The two-letter abbreviation for the airline.
- **OriginAirportID**: A unique numeric identifier for the departure aiport
- **OriginAirportName**: The full name of the departure airport
- **OriginCity**: The departure airport city
- **OriginState**: The departure airport state
- **DestAirportID**: A unique numeric identifier for the destination aiport
- **DestAirportName**: The full name of the destination airport
- **DestCity**: The destination airport city
- **DestState**: The destination airport state
- **CRSDepTime**: The scheduled departure time
- **DepDelay**: The number of minutes departure was delayed (flight that left ahead of schedule have a negative value)
- **DelDelay15**: A binary indicator that departure was delayed by more than 15 minutes (and therefore considered "late")
- **CRSArrTime**: The scheduled arrival time
- **ArrDelay**: The number of minutes arrival was delayed (flight that arrived ahead of schedule have a negative value)
- **ArrDelay15**: A binary indicator that arrival was delayed by more than 15 minutes (and therefore considered "late")
- **Cancelled**: A binary indicator that the flight was cancelled

**2. Identify and eliminate any outliers in the DepDelay and ArrDelay columns.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,axes=plt.subplots(2,1,figsize=(10,6))
sns.boxplot(x=df_flights['DepDelay'],ax=axes[0])
axes[1].hist(x=df_flights['DepDelay'])
plt.xlabel('DepDelay')
```

```python
fig,axes=plt.subplots(2,1,figsize=(10,6))
sns.boxplot(x=df_flights['ArrDelay'],ax=axes[0])
axes[1].hist(x=df_flights['ArrDelay'])
plt.xlabel('ArrDelay')
```

```python
df_flights[['DepDelay','ArrDelay']].describe().transpose()
```

```python
import numpy as np
## DepDelay outliers
## Since the left skewed 
# so min-value and max value for the outlier is
DepDelayMinMax=list(df_flights.DepDelay.quantile([0.01,0.9]))
ArrDelayMinMax=list(df_flights.ArrDelay.quantile([0.05,0.9]))
# # Deleting outliers
df_flights=df_flights[np.logical_and(df_flights.DepDelay>DepDelayMinMax[0], df_flights.DepDelay <DepDelayMinMax[1])]
df_flights=df_flights[np.logical_and(df_flights.ArrDelay>ArrDelayMinMax[0] , df_flights.ArrDelay<ArrDelayMinMax[1])]
df_flights.shape
```

```python
fig,axes=plt.subplots(2,1,figsize=(10,6))
sns.boxplot(x=df_flights['DepDelay'],ax=axes[0])
axes[1].hist(x=df_flights['DepDelay'])
plt.xlabel('DepDelay')
```

```python
fig,axes=plt.subplots(2,1,figsize=(10,6))
sns.boxplot(x=df_flights['ArrDelay'],ax=axes[0])
axes[1].hist(x=df_flights['ArrDelay'])
plt.xlabel('ArrDelay')
```

After removing the outliers the distribution of DepDelay and ArrDelay is much better

## Use statistics, aggregate functions, and visualizations to answer the following questions:

What are the average (mean) departure and arrival delays?

```python
df_flights[['DepDelay','ArrDelay']].mean()
```

How do the carriers compare in terms of arrival delay performance?




```python
fig,axes=plt.subplots(1,2,figsize=(18,6))

sns.boxplot(x='Carrier',y='ArrDelay',data=df_flights,ax=axes[0]).set(title='Carriers compare in terms of arrival performance')
sns.boxplot(x='Carrier',y='DepDelay',data=df_flights,ax=axes[1]).set(title='Carriers compare in terms of Departure performance')
plt.show()
```

Is there a noticable difference in arrival delays for different days of the week?

```python
fig,axes=plt.subplots(1,2,figsize=(18,6))

sns.boxplot(x='DayOfWeek',y='ArrDelay',data=df_flights,ax=axes[0])
sns.boxplot(x='DayOfWeek',y='DepDelay',data=df_flights,ax=axes[1])
plt.show()
```

Which departure airport has the highest average departure delay?

```python
print(df_flights.groupby('OriginAirportName').agg({'DepDelay':'mean'}).sort_values(by='DepDelay',ascending=False))
df_flights.groupby('OriginAirportName').agg({'DepDelay':'mean'}).sort_values(by='DepDelay',ascending=False).plot(kind='bar',figsize=(11,10))
plt.title('Airport vs Average Departure Delay')
```

Do late departures tend to result in longer arrival delays than on-time departures?

```python
sns.boxplot(x='DepDel15',y='ArrDelay',data=df_flights)
```

Which route (from origin airport to destination airport) has the most late arrivals?

```python
df_flights['Route']=df_flights.apply(lambda x :'('+ x['OriginAirportName'] +') to ('+ x['DestAirportName']+')',axis=1)
```

```python
df_flights['Route']
```

```python
df_flights.groupby('Route').agg({'ArrDel15':'sum'}).sort_values(by='ArrDel15',ascending=False)
```

Which route has the highest average arrival delay?

```python
df_flights.groupby('Route').agg({'ArrDelay':'mean'}).sort_values(by='ArrDelay',ascending=False)
```
